```python
import subprocess, sys

check = "from transformers import TrainingArguments; from trl import GRPOConfig, GRPOTrainer, SFTConfig, SFTTrainer; print('training stack ok')"
# torchvision is not needed for this text-only training job. On HF Spaces it can
# be preinstalled with a CUDA build that does not match torch, which breaks
# Transformers imports before training even starts.
subprocess.run([sys.executable, "-m", "pip", "uninstall", "-y", "torchvision"], check=False)
try:
    subprocess.check_call([sys.executable, "-c", check])
except subprocess.CalledProcessError:
    print('Installing/fixing the text training stack...')

    packages = [
        "huggingface-hub==0.36.0",
        "transformers==4.56.2",
        "trl==0.24.0",
        "accelerate==1.10.1",
        "peft==0.17.1",
        "datasets==4.3.0",
        "mergekit==0.1.4",
    ]
    subprocess.check_call([sys.executable, "-m", "pip", "install", "--upgrade", "--prefer-binary", "--no-deps", *packages])
    subprocess.check_call([sys.executable, "-c", check])
```
